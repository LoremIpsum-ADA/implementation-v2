```python
# 4b. GBIF bulk occurrence download workflow
from pygbif import occurrences as occ
import time
import zipfile
import os

def request_gbif_download(species_list, year_min, year_max, bbox, user, pwd, email):
    """
    Request bulk occurrence download from GBIF.
    Returns download key for later retrieval.
    """
    if not all([user, pwd, email]):
        print("ERROR: GBIF credentials required for bulk download")
        print("Set CONFIG['gbif_user'], CONFIG['gbif_password'], CONFIG['gbif_email']")
        return None

    # Get taxon keys for species
    taxon_keys = []
    for sp_name in species_list:
        try:
            result = gbif_species.name_backbone(name=sp_name)
            if result.get('usageKey'):
                taxon_keys.append(result['usageKey'])
        except:
            pass

    if not taxon_keys:
        print("No valid taxon keys found for species list")
        return None

    print(f"Found {len(taxon_keys)} taxon keys for {len(species_list)} species")

    # Build download predicate using proper GBIF predicate format
    # Create OR predicate for taxon keys
    taxon_predicates = [{"type": "equals", "key": "TAXON_KEY", "value": str(key)} for key in taxon_keys]

    # Build complete predicate
    predicate = {
        "type": "and",
        "predicates": [
            # Taxon keys (OR'd together)
            {
                "type": "or",
                "predicates": taxon_predicates
            },
            # Has coordinates
            {"type": "equals", "key": "HAS_COORDINATE", "value": "true"},
            # No geospatial issues
            {"type": "equals", "key": "HAS_GEOSPATIAL_ISSUE", "value": "false"},
            # Year range
            {"type": "greaterThanOrEquals", "key": "YEAR", "value": str(year_min)},
            {"type": "lessThanOrEquals", "key": "YEAR", "value": str(year_max)},
            # Geographic bounding box - use GEOMETRY predicate
            {
                "type": "within",
                "geometry": f"POLYGON(({bbox[0]} {bbox[1]},{bbox[2]} {bbox[1]},{bbox[2]} {bbox[3]},{bbox[0]} {bbox[3]},{bbox[0]} {bbox[1]}))"
            }
        ]
    }

    try:
        # occ.download returns a tuple: (download_key, created_date)
        result = occ.download(predicate, user=user, pwd=pwd, email=email)

        # Extract download key from tuple
        if isinstance(result, tuple):
            download_key = result[0]
        else:
            download_key = result

        print(f"Download requested. Key: {download_key}")
        print(f"You will receive an email at {email} when ready")
        print(f"Download URL: https://www.gbif.org/occurrence/download/{download_key}")

        return download_key

    except Exception as e:
        print(f"Error requesting download: {e}")
        import traceback
        traceback.print_exc()
        print("\nAlternative: Use GBIF web interface:")
        print("1. Go to https://www.gbif.org/occurrence/search")
        print("2. Filter by species, date range, and coordinates")
        print("3. Click 'Download' and select 'Simple CSV'")
        print(f"4. Save to {CONFIG['data_dir'] / 'raw' / 'gbif' / 'occurrences.csv'}")
        return None

# ---- NOT WORKING, DOWNLOAD MANUALLY ----
# def download_gbif_data(download_key, user, pwd, output_dir):
#     """
#     Download and extract GBIF occurrence data once ready.
#     """
#     if not download_key:
#         print("No download key provided")
#         return None

#     output_dir = Path(output_dir)
#     output_dir.mkdir(parents=True, exist_ok=True)

#     try:
#         # Check download status
#         metadata = occ.download_meta(download_key)
#         status = metadata.get('status', 'UNKNOWN')

#         print(f"Download status: {status}")

#         if status != 'SUCCEEDED':
#             print(f"Download not ready yet. Current status: {status}")
#             print("Please wait for email notification or check status later")
#             return None

#         # Download the file
#         print("Downloading data...")

#         # occ.download_get returns a tuple: (download_key, path)
#         result = occ.download_get(download_key, path=str(output_dir))

#         # Extract the path from the tuple
#         if isinstance(result, tuple):
#             _, download_path = result
#         else:
#             download_path = result

#         zip_path = Path(download_path)

#         # Extract
#         print("Extracting data...")
#         with zipfile.ZipFile(zip_path, 'r') as zip_ref:
#             zip_ref.extractall(output_dir)

#         # Find the occurrence file
#         occurrence_file = output_dir / 'occurrence.txt'
#         if occurrence_file.exists():
#             # Rename to CSV
#             output_csv = output_dir / 'occurrences.csv'
#             occurrence_file.rename(output_csv)
#             print(f"Success! Data saved to {output_csv}")
#             return output_csv
#         else:
#             print("Could not find occurrence.txt in download")
#             return None

#     except Exception as e:
#         print(f"Error downloading: {e}")
#         import traceback
#         traceback.print_exc()
#         return None

# Example usage
if CONFIG['gbif_email']:
    print("GBIF Download Workflow:")
    print("=" * 60)

    # Get species names
    if 'species_df' in locals() and not species_df.empty:
        species_names = species_df['scientificName'].tolist()

        print(f"\nStep 1: Request download for {len(species_names)} species")
        print("Run the cell below to initiate download")
        print("\nStep 2: After receiving email, download data and put it into data/raw/gbif/")
    else:
        print("First run cell 4a to get species list")
else:
    print("Please set CONFIG['gbif_email'] to use GBIF download workflow")
    print("\nAlternative: Use GBIF web interface and place CSV in data/raw/gbif/")

gbif_raw_path = CONFIG['data_dir'] / 'raw' / 'gbif' / 'occurrences.csv'
print(f"\nExpected GBIF data path: {gbif_raw_path}")
```

```python
# 4c. EM-DAT disaster data ingestion
import pandas as pd

def load_emdat_data(filepath, disaster_types, year_min, year_max, bbox):
    """
    Load and filter EM-DAT disaster data.
    Expected columns from EM-DAT: Start Year, Disaster Type, Country, Latitude, Longitude, etc.
    """
    if not filepath.exists():
        print(f"ERROR: EM-DAT file not found at {filepath}")
        print("Download from https://www.emdat.be/ and place in data/raw/")
        return None

    # Load with encoding that handles special characters
    df = pd.read_csv(filepath, encoding='utf-8-sig', low_memory=False)

    # Print column names for debugging
    print("Available columns:", df.columns.tolist()[:10], "...")

    # The actual EM-DAT file uses 'Start Year' not 'year'
    # and 'Disaster Type' not 'disaster_type'

    # Filter by year - use 'Start Year' column
    if 'Start Year' in df.columns:
        df = df[(df['Start Year'] >= year_min) & (df['Start Year'] <= year_max)]
    else:
        print("WARNING: 'Start Year' column not found")
        return None

    # Filter by disaster type - map to EM-DAT disaster types
    disaster_map = {
        'wildfire': ['Wildfire'],
        'flood': ['Flood', 'Flash flood', 'Riverine flood', 'Coastal flood'],
        'cyclone': ['Storm', 'Tropical cyclone', 'Typhoon', 'Cyclone'],
        'earthquake': ['Earthquake', 'Ground movement']
    }

    type_filter = []
    for dtype in disaster_types:
        if dtype in disaster_map:
            type_filter.extend(disaster_map[dtype])

    # Use 'Disaster Subtype' or 'Disaster Type' column
    if 'Disaster Subtype' in df.columns:
        df = df[df['Disaster Subtype'].isin(type_filter)]
    elif 'Disaster Type' in df.columns:
        df = df[df['Disaster Type'].isin(type_filter)]
    else:
        print("WARNING: Disaster type columns not found")

    # Filter by geography (if coordinates available)
    if 'Latitude' in df.columns and 'Longitude' in df.columns:
        # Remove rows with missing coordinates
        df = df.dropna(subset=['Latitude', 'Longitude'])
        df = df[
            (df['Longitude'] >= bbox[0]) & (df['Longitude'] <= bbox[2]) &
            (df['Latitude'] >= bbox[1]) & (df['Latitude'] <= bbox[3])
        ]
    else:
        print("WARNING: Latitude/Longitude columns not found - filtering by country names instead")
        # Fallback: filter by Asian countries if coordinates not available
        # You may need to expand this list
        asian_countries = ['China', 'India', 'Indonesia', 'Pakistan', 'Bangladesh',
                          'Japan', 'Philippines', 'Vietnam', 'Thailand', 'Myanmar',
                          'Korea', 'Nepal', 'Sri Lanka', 'Malaysia', 'Cambodia']
        if 'Country' in df.columns:
            df = df[df['Country'].isin(asian_countries)]

    # Standardize column names for downstream use
    column_mapping = {
        'Start Year': 'year',
        'Disaster Type': 'disaster_type',
        'Disaster Subtype': 'disaster_subtype',
        'Latitude': 'latitude',
        'Longitude': 'longitude',
        'Total Deaths': 'total_deaths',
        'No. Affected': 'affected',
        'Total Damage (\'000 US$)': 'damage_usd'
    }

    # Rename columns that exist
    rename_dict = {k: v for k, v in column_mapping.items() if k in df.columns}
    df = df.rename(columns=rename_dict)

    print(f"Loaded {len(df)} disaster events from {year_min} to {year_max}")
    return df

emdat_path = CONFIG['data_dir'] / 'raw' / 'emdat_asia_2000_2024.csv'
emdat_df = load_emdat_data(
    emdat_path,
    CONFIG['disaster_types'],
    CONFIG['time_range'][0],
    CONFIG['time_range'][1],
    CONFIG['region_bbox']
)

if emdat_df is not None:
    print("\nEM-DAT data summary:")
    if 'disaster_type' in emdat_df.columns:
        print(emdat_df.groupby('disaster_type').size())
    elif 'disaster_subtype' in emdat_df.columns:
        print(emdat_df.groupby('disaster_subtype').size())

    print(f"\nYears covered: {emdat_df['year'].min()} to {emdat_df['year'].max()}")
    print(f"Events with coordinates: {emdat_df[['latitude', 'longitude']].notna().all(axis=1).sum()}")
```

```python
# 4d. IUCN range map processing
import geopandas as gpd
from pathlib import Path

def load_iucn_ranges(iucn_dir, species_list):
    """
    Load IUCN range shapefiles for species in species_list.
    Handles PART1/PART2 shapefiles for mammals/reptiles/amphibians and GPKG for birds.
    """
    if not iucn_dir.exists():
        print(f"ERROR: IUCN directory not found at {iucn_dir}")
        print("Download from https://www.iucnredlist.org/resources/spatial-data-download")
        return None

    gdfs = []

    # Load MAMMALS (PART1 and PART2 shapefiles)
    mammals_dir = iucn_dir / 'MAMMALS'
    if mammals_dir.exists():
        print("Loading MAMMALS data...")
        for part in ['PART1', 'PART2']:
            shp_file = mammals_dir / f'MAMMALS_{part}.shp'
            if shp_file.exists():
                print(f"  Loading MAMMALS_{part}.shp...")
                gdf = gpd.read_file(shp_file)
                # Filter to species of interest
                if 'binomial' in gdf.columns:
                    gdf = gdf[gdf['binomial'].isin(species_list)]
                    gdfs.append(gdf)
                elif 'sci_name' in gdf.columns:
                    gdf = gdf[gdf['sci_name'].isin(species_list)]
                    gdfs.append(gdf)

    # Load BIRDS (GPKG format)
    birds_dir = iucn_dir / 'BIRDS'
    if birds_dir.exists():
        gpkg_file = birds_dir / 'BOTW_2025.gpkg'
        if gpkg_file.exists():
            print("Loading BIRDS data from GPKG...")
            gdf = gpd.read_file(gpkg_file)
            # Filter to species of interest - birds typically use 'sci_name' or 'SCINAME'
            name_col = None
            for col in ['sci_name', 'SCINAME', 'binomial', 'scientific_name']:
                if col in gdf.columns:
                    name_col = col
                    break
            if name_col:
                gdf = gdf[gdf[name_col].isin(species_list)]
                gdfs.append(gdf)

    # Load REPTILES (PART1 and PART2 shapefiles)
    reptiles_dir = iucn_dir / 'REPTILES'
    if reptiles_dir.exists():
        print("Loading REPTILES data...")
        for part in ['PART1', 'PART2']:
            shp_file = reptiles_dir / f'REPTILES_{part}.shp'
            if shp_file.exists():
                print(f"  Loading REPTILES_{part}.shp...")
                gdf = gpd.read_file(shp_file)
                # Filter to species of interest
                if 'binomial' in gdf.columns:
                    gdf = gdf[gdf['binomial'].isin(species_list)]
                    gdfs.append(gdf)
                elif 'sci_name' in gdf.columns:
                    gdf = gdf[gdf['sci_name'].isin(species_list)]
                    gdfs.append(gdf)

    # Load AMPHIBIANS (PART1 and PART2 shapefiles)
    amphibians_dir = iucn_dir / 'AMPHIBIANS'
    if amphibians_dir.exists():
        print("Loading AMPHIBIANS data...")
        for part in ['PART1', 'PART2']:
            shp_file = amphibians_dir / f'AMPHIBIANS_{part}.shp'
            if shp_file.exists():
                print(f"  Loading AMPHIBIANS_{part}.shp...")
                gdf = gpd.read_file(shp_file)
                # Filter to species of interest
                if 'binomial' in gdf.columns:
                    gdf = gdf[gdf['binomial'].isin(species_list)]
                    gdfs.append(gdf)
                elif 'sci_name' in gdf.columns:
                    gdf = gdf[gdf['sci_name'].isin(species_list)]
                    gdfs.append(gdf)

    if not gdfs:
        print("No matching species found in IUCN data")
        return None

    # Combine all GeoDataFrames
    ranges = pd.concat(gdfs, ignore_index=True)

    # Standardize column names
    if 'sci_name' in ranges.columns and 'binomial' not in ranges.columns:
        ranges['binomial'] = ranges['sci_name']

    print(f"Loaded {len(ranges)} range polygons for {ranges['binomial'].nunique()} species")

    return ranges

iucn_dir = CONFIG['data_dir'] / 'raw' / 'iucn'
species_names = species_df['scientificName'].tolist() if 'species_df' in locals() else []

iucn_ranges = load_iucn_ranges(iucn_dir, species_names)
```

```python
# 5. Create spatial analysis grid
import numpy as np
import geopandas as gpd
from shapely.geometry import box

def create_analysis_grid(bbox, resolution):
    """
    Create regular lat-lon grid covering bounding box.
    Returns GeoDataFrame with grid cells.
    """
    min_lon, min_lat, max_lon, max_lat = bbox

    lons = np.arange(min_lon, max_lon, resolution)
    lats = np.arange(min_lat, max_lat, resolution)

    grid_polys = []
    grid_ids = []
    grid_centers = []

    for i, lon in enumerate(lons):
        for j, lat in enumerate(lats):
            poly = box(lon, lat, lon + resolution, lat + resolution)
            grid_polys.append(poly)
            grid_ids.append(f'g_{i}_{j}')
            grid_centers.append((lon + resolution/2, lat + resolution/2))

    grid_gdf = gpd.GeoDataFrame({
        'grid_id': grid_ids,
        'lon_center': [c[0] for c in grid_centers],
        'lat_center': [c[1] for c in grid_centers],
        'geometry': grid_polys
    }, crs='EPSG:4326')

    # Calculate grid area in km²
    grid_gdf['area_km2'] = grid_gdf.to_crs('EPSG:3857').geometry.area / 1e6

    return grid_gdf

grid_gdf = create_analysis_grid(CONFIG['region_bbox'], CONFIG['grid_resolution_deg'])
print(f"Created grid with {len(grid_gdf):,} cells")
print(f"Avg cell area: {grid_gdf['area_km2'].mean():.1f} km²")

# Save grid for reference
grid_path = CONFIG['data_dir'] / 'processed' / 'analysis_grid.gpkg'
grid_gdf.to_file(grid_path, driver='GPKG')
print(f"Saved grid to {grid_path}")

grid_gdf.head()
```

```python
# 6. Spatial join functions - link occurrences and disasters to grid
import geopandas as gpd

def map_points_to_grid(points_gdf, grid_gdf):
    """
    Spatially join point data to grid cells.
    """
    # Ensure same CRS
    pts = points_gdf.to_crs(grid_gdf.crs)

    # Spatial join
    joined = gpd.sjoin(pts, grid_gdf[['grid_id', 'geometry']], how='left', predicate='within')

    # Drop points outside grid
    joined = joined.dropna(subset=['grid_id'])

    return joined
def map_polygon_to_grid(poly_gdf, grid_gdf):
    """
    Compute grid cell overlap with polygon ranges.
    Returns grid cells with fraction of area covered.
    """
    poly = poly_gdf.to_crs(grid_gdf.crs)

    # Intersection
    intersected = gpd.overlay(poly, grid_gdf[['grid_id', 'geometry']], how='intersection')

    # Calculate overlap area
    intersected['overlap_km2'] = intersected.to_crs('EPSG:3857').geometry.area / 1e6

    # Merge with grid areas to get fraction
    intersected = intersected.merge(
        grid_gdf[['grid_id', 'area_km2']],
        on='grid_id',
        how='left'
    )
    intersected['overlap_fraction'] = intersected['overlap_km2'] / intersected['area_km2']

    return intersected

def expand_disaster_footprint(disaster_points, radius_km, grid_gdf):
    """
    Expand point disasters to circular footprints.
    Returns affected grid cells.
    """
    from shapely.geometry import Point

    # Convert to projected CRS for buffering
    pts = disaster_points.to_crs('EPSG:3857')

    # Buffer by radius
    pts['geometry'] = pts.geometry.buffer(radius_km * 1000)

    # Back to WGS84
    pts = pts.to_crs('EPSG:4326')

    # Intersect with grid
    affected = map_polygon_to_grid(pts, grid_gdf)

    return affected

print("Spatial join functions defined")
```

```python
# 7. Occurrence data processing and outcome construction
import pandas as pd

def load_and_process_gbif_occurrences(filepath, species_list, year_min, year_max, bbox):
    """
    Load GBIF occurrence CSV and prepare for analysis.
    Note: GBIF exports are tab-separated even when saved as .csv
    """
    if not filepath.exists():
        print(f"ERROR: GBIF file not found at {filepath}")
        return None

    # GBIF data is tab-separated, not comma-separated
    # Try to detect separator automatically
    try:
        # First, try reading as tab-separated (GBIF default)
        df = pd.read_csv(filepath, sep='\t', encoding='utf-8',engine='python', quoting=3)

        # Check if we got valid columns - GBIF files should have many columns
        if len(df.columns) < 10:
            # If we only got a few columns, file might be comma-separated
            print("Tab separation failed, trying comma separator...")
            df = pd.read_csv(filepath, sep=',', low_memory=False, encoding='utf-8')
    except Exception as e:
        print(f"Error loading file: {e}")
        return None

    print(f"Loaded file with {len(df)} rows and {len(df.columns)} columns")

    # Standardize columns
    required_cols = ['species', 'decimalLatitude', 'decimalLongitude', 'year']
    missing_cols = [col for col in required_cols if col not in df.columns]

    if missing_cols:
        print(f"ERROR: Missing required columns: {missing_cols}")
        print(f"Available columns: {df.columns.tolist()[:20]}...")
        return None

    # Filter by species
    df = df[df['species'].isin(species_list)]
    print(f"After species filter: {len(df)} records")

    # Filter by year
    df['year'] = pd.to_numeric(df['year'], errors='coerce')
    df = df[(df['year'] >= year_min) & (df['year'] <= year_max)]
    print(f"After year filter ({year_min}-{year_max}): {len(df)} records")

    # Filter by coordinates
    df = df[
        (df['decimalLongitude'] >= bbox[0]) & (df['decimalLongitude'] <= bbox[2]) &
        (df['decimalLatitude'] >= bbox[1]) & (df['decimalLatitude'] <= bbox[3])
    ]
    print(f"After bbox filter: {len(df)} records")

    # Drop invalid coordinates
    df = df.dropna(subset=['decimalLatitude', 'decimalLongitude'])

    print(f"\nFinal: {len(df):,} occurrences for {df['species'].nunique()} species")

    return df

def build_occurrence_panel(occurrence_df, grid_gdf, time_unit='year'):
    """
    Create grid × time × species panel with occurrence counts.
    """
    # Convert to GeoDataFrame
    occ_gdf = gpd.GeoDataFrame(
        occurrence_df,
        geometry=gpd.points_from_xy(
            occurrence_df['decimalLongitude'],
            occurrence_df['decimalLatitude']
        ),
        crs='EPSG:4326'
    )

    # Map to grid
    occ_joined = map_points_to_grid(occ_gdf, grid_gdf)

    # Aggregate by grid × time × species
    agg = occ_joined.groupby(['grid_id', time_unit, 'species']).size().reset_index(name='n_occurrences')

    # Add effort proxy (number of unique collection events)
    if 'eventID' in occ_joined.columns:
        effort = occ_joined.groupby(['grid_id', time_unit])['eventID'].nunique().reset_index(name='n_events')
        agg = agg.merge(effort, on=['grid_id', time_unit], how='left')

    # Calculate detection-adjusted occupancy
    agg['occupancy'] = (agg['n_occurrences'] > 0).astype(int)

    return agg

# Example usage: Load GBIF data if available
gbif_path = CONFIG['data_dir'] / 'raw' / 'gbif' / '0019239-251025141854904.csv'

if gbif_path.exists():
    print(f"Found GBIF data at {gbif_path}")
    print("Loading and processing occurrence data...")

    # Get species list from previous step
    if 'species_df' in locals() and not species_df.empty:
        species_names = species_df['scientificName'].tolist()

        # Load and process occurrences
        occurrence_df = load_and_process_gbif_occurrences(
            gbif_path,
            species_names,
            CONFIG['time_range'][0],
            CONFIG['time_range'][1],
            CONFIG['region_bbox']
        )

        if occurrence_df is not None and len(occurrence_df) > 0:
            # Build occurrence panel
            if 'grid_gdf' in locals():
                occurrence_panel = build_occurrence_panel(occurrence_df, grid_gdf, time_unit='year')

                print(f"\nOccurrence panel created:")
                print(f"  Grid cells: {occurrence_panel['grid_id'].nunique():,}")
                print(f"  Species: {occurrence_panel['species'].nunique()}")
                print(f"  Years: {occurrence_panel['year'].min()} - {occurrence_panel['year'].max()}")
                print(f"  Total records: {len(occurrence_panel):,}")

                # Save panel
                panel_path = CONFIG['data_dir'] / 'processed' / 'occurrence_panel.csv'
                occurrence_panel.to_csv(panel_path, index=False)
                print(f"\nSaved occurrence panel to {panel_path}")

                # Display summary statistics
                print("\nOccurrence summary by species:")
                species_summary = occurrence_panel.groupby('species').agg({
                    'n_occurrences': 'sum',
                    'grid_id': 'nunique'
                }).rename(columns={'grid_id': 'n_grid_cells'}).sort_values('n_occurrences', ascending=False)
                print(species_summary.head(10))
            else:
                print("ERROR: Grid not found. Run cell 5 to create analysis grid first.")
        else:
            print("No occurrence data loaded. Check GBIF file format.")
    else:
        print("Species list not found. Run cell 4a to get species list first.")
else:
    print(f"GBIF data not found at {gbif_path}")
    print("\nTo proceed:")
    print("1. Run cells 4a and 4b to request GBIF download")
    print("2. Wait for email notification from GBIF")
    print("3. Place downloaded occurrence file at:")
    print(f"   {gbif_path}")
    print("4. Re-run this cell to process the data")
```

```python
# 7b. BUILD BALANCED OCCURRENCE PANEL (CRITICAL FIX)
import itertools
import numpy as np  # Added missing import
import pandas as pd
import geopandas as gpd

def build_occurrence_panel_BALANCED(occurrence_df, grid_gdf, species_list, years, time_unit='year'):
    """
    CRITICAL FIX: Create BALANCED panel with zeros for unobserved grid-species-year combinations.

    This fixes the "float division by zero" error by ensuring we have both 0s and 1s in occupancy.
    """
    print("Building BALANCED occurrence panel...")

    # Convert to GeoDataFrame
    occ_gdf = gpd.GeoDataFrame(
        occurrence_df,
        geometry=gpd.points_from_xy(
            occurrence_df['decimalLongitude'],
            occurrence_df['decimalLatitude']
        ),
        crs='EPSG:4326'
    )

    # Map occurrences to grid
    occ_joined = map_points_to_grid(occ_gdf, grid_gdf)

    # Aggregate actual observations
    observed = occ_joined.groupby(['grid_id', time_unit, 'species']).agg({
        'decimalLatitude': 'count'
    }).reset_index()
    observed.columns = ['grid_id', time_unit, 'species', 'n_occurrences']
    observed['occupancy'] = 1

    print(f"  Observed records: {len(observed):,}")

    # CRITICAL: Create full balanced panel
    years_list = list(range(years[0], years[1] + 1))
    grid_ids = grid_gdf['grid_id'].unique()

    # All possible combinations
    all_combinations = pd.DataFrame(
        list(itertools.product(grid_ids, years_list, species_list)),
        columns=['grid_id', time_unit, 'species']
    )

    print(f"  Total possible combinations: {len(all_combinations):,}")

    # Merge with observed data
    balanced = all_combinations.merge(
        observed,
        on=['grid_id', time_unit, 'species'],
        how='left'
    )

    # Fill missing values
    balanced['n_occurrences'] = balanced['n_occurrences'].fillna(0)
    balanced['occupancy'] = balanced['occupancy'].fillna(0)

    # Keep only grid cells where species CAN occur (using IUCN ranges)
    print("  Filtering to species ranges...")

    # Optional: Filter to reduce size while maintaining balance
    if 'iucn_ranges' in globals():
        valid_grid_species = get_valid_grid_species_combinations(grid_gdf, iucn_ranges)
        balanced = balanced.merge(valid_grid_species, on=['grid_id', 'species'], how='inner')

    print(f"  Final balanced panel: {len(balanced):,} rows")
    print(f"  Occupancy = 1: {balanced['occupancy'].sum():,} ({100*balanced['occupancy'].mean():.2f}%)")
    print(f"  Occupancy = 0: {(balanced['occupancy']==0).sum():,} ({100*(1-balanced['occupancy'].mean()):.2f}%)")

    # CRITICAL CHECK
    if balanced['occupancy'].std() == 0:
        raise ValueError("ERROR: Occupancy has zero variance! Panel construction failed.")

    return balanced

def get_valid_grid_species_combinations(grid_gdf, iucn_ranges):
    """
    Reduce panel size by only keeping grid cells within species ranges.
    This maintains balance while reducing computational burden.
    """
    valid_combos = []

    for species in iucn_ranges['binomial'].unique():
        # Get range for this species
        species_range = iucn_ranges[iucn_ranges['binomial'] == species]

        # Spatial join to find overlapping grid cells
        overlapping = gpd.sjoin(
            grid_gdf[['grid_id', 'geometry']],
            species_range[['geometry']],
            how='inner',
            predicate='intersects'
        )

        overlapping['species'] = species
        valid_combos.append(overlapping[['grid_id', 'species']])

    if valid_combos:
        result = pd.concat(valid_combos, ignore_index=True).drop_duplicates()
        print(f"  Valid grid-species combinations: {len(result):,}")
        return result
    else:
        # Fallback: all combinations
        return pd.DataFrame({
            'grid_id': grid_gdf['grid_id'].repeat(len(iucn_ranges['binomial'].unique())),
            'species': np.tile(iucn_ranges['binomial'].unique(), len(grid_gdf))
        })

# Now rebuild the occurrence panel using the BALANCED method
if 'occurrence_df' in locals() and 'grid_gdf' in locals() and 'species_df' in locals():
    print("\n" + "="*80)
    print("REBUILDING OCCURRENCE PANEL WITH BALANCED METHOD (CRITICAL FIX)")
    print("="*80)

    species_names = species_df['scientificName'].tolist()

    occurrence_panel = build_occurrence_panel_BALANCED(
        occurrence_df,
        grid_gdf,
        species_names,
        CONFIG['time_range'],
        time_unit='year'
    )

    print(f"\n✓ Balanced occurrence panel created successfully!")
    print(f"  Grid cells: {occurrence_panel['grid_id'].nunique():,}")
    print(f"  Species: {occurrence_panel['species'].nunique()}")
    print(f"  Years: {occurrence_panel['year'].min()} - {occurrence_panel['year'].max()}")
    print(f"  Total records: {len(occurrence_panel):,}")
    print(f"  Occupancy variance: {occurrence_panel['occupancy'].std():.6f} (MUST BE > 0)")

    # Save the balanced panel
    panel_path = CONFIG['data_dir'] / 'processed' / 'occurrence_panel_balanced.csv'
    occurrence_panel.to_csv(panel_path, index=False)
    print(f"\n✓ Saved to {panel_path}")
else:
    print("ERROR: Missing required data. Ensure occurrence_df, grid_gdf, and species_df are loaded.")
```

```python
# 8. Treatment panel construction
import pandas as pd
import geopandas as gpd

def build_treatment_panel(disaster_df, grid_gdf, time_unit='year', buffer_km=50):
    """
    Create treatment indicators for grid × time.

    Parameters:
    - disaster_df: DataFrame with columns [year, latitude, longitude, disaster_type, ...]
    - grid_gdf: Grid GeoDataFrame
    - buffer_km: radius to expand point events (default 50km)

    Returns: DataFrame with [grid_id, year, treated, disaster_type, intensity]
    """
    if disaster_df is None or len(disaster_df) == 0:
        print("No disaster data provided")
        return None

    # Convert to GeoDataFrame
    disaster_gdf = gpd.GeoDataFrame(
        disaster_df,
        geometry=gpd.points_from_xy(disaster_df['longitude'], disaster_df['latitude']),
        crs='EPSG:4326'
    )

    # Expand to footprints
    affected = expand_disaster_footprint(disaster_gdf, buffer_km, grid_gdf)

    # Aggregate by grid × time
    treatment = affected.groupby(['grid_id', time_unit]).agg({
        'disaster_type': 'first',  # Take first if multiple
        'overlap_fraction': 'sum'  # Sum overlaps (can be >1 if multiple events)
    }).reset_index()

    # Create binary treatment
    treatment['treated'] = 1
    treatment['treatment_intensity'] = treatment['overlap_fraction']

    # Create event timing for event study
    # For each grid, find first treatment year
    first_treatment = treatment.groupby('grid_id')[time_unit].min().reset_index()
    first_treatment.columns = ['grid_id', 'first_treatment_year']
    treatment = treatment.merge(first_treatment, on='grid_id', how='left')

    return treatment

def create_full_panel(grid_gdf, years, treatment_df=None):
    """
    Create complete grid × year panel with treatment indicators.
    """
    # All grid × year combinations
    years_list = list(range(years[0], years[1] + 1))
    panel = pd.DataFrame([
        {'grid_id': gid, 'year': yr}
        for gid in grid_gdf['grid_id']
        for yr in years_list
    ])

    # Merge treatment
    if treatment_df is not None:
        panel = panel.merge(
            treatment_df[['grid_id', 'year', 'treated', 'treatment_intensity', 'first_treatment_year']],
            on=['grid_id', 'year'],
            how='left'
        )
        panel['treated'] = panel['treated'].fillna(0)
        panel['treatment_intensity'] = panel['treatment_intensity'].fillna(0)
    else:
        panel['treated'] = 0
        panel['treatment_intensity'] = 0

    return panel

print("Treatment construction functions defined")
```

```python
# # 9. Difference-in-Differences estimation (FIXED VERSION)
# import pandas as pd
# import numpy as np
# from linearmodels.panel import PanelOLS
# from linearmodels import IV2SLS
# import statsmodels.api as sm
# from statsmodels.regression.linear_model import OLS

# def prepare_panel_data(outcome_df, treatment_df, grid_gdf):
#     """
#     Merge outcome and treatment into analysis panel.
#     """
#     # Merge treatment with outcomes
#     panel = outcome_df.merge(
#         treatment_df[['grid_id', 'year', 'treated', 'treatment_intensity', 'first_treatment_year']],
#         on=['grid_id', 'year'],
#         how='left'
#     )

#     # Fill missing treatment as control
#     panel['treated'] = panel['treated'].fillna(0)
#     panel['treatment_intensity'] = panel['treatment_intensity'].fillna(0)

#     # Add grid characteristics
#     panel = panel.merge(
#         grid_gdf[['grid_id', 'lon_center', 'lat_center', 'area_km2']],
#         on='grid_id',
#         how='left'
#     )

#     return panel

# def estimate_twfe_did_FIXED(panel_df, outcome_var, treatment_var='treated',
#                             entity_var='grid_id', time_var='year'):
#     """
#     FIXED: Two-way fixed effects DiD with proper error handling.
#     """
#     # Validate inputs
#     if panel_df[outcome_var].std() == 0:
#         print("ERROR: Outcome variable has zero variance!")
#         print(f"Unique values: {panel_df[outcome_var].unique()}")
#         return None

#     if panel_df[treatment_var].std() == 0:
#         print("WARNING: Treatment variable has zero variance!")
#         print(f"Unique values: {panel_df[treatment_var].unique()}")
#         return None

#     # Check for sufficient variation
#     n_treated = panel_df[treatment_var].sum()
#     n_total = len(panel_df)
#     print(f"Treatment prevalence: {n_treated}/{n_total} ({100*n_treated/n_total:.2f}%)")

#     # Set multi-index for panel
#     panel = panel_df.copy()
#     panel = panel.set_index([entity_var, time_var])

#     # Build formula
#     formula = f'{outcome_var} ~ {treatment_var} + EntityEffects + TimeEffects'

#     try:
#         mod = PanelOLS.from_formula(formula, data=panel)
#         res = mod.fit(cov_type='clustered', cluster_entity=True)

#         print("\nTwo-Way Fixed Effects DiD Results")
#         print("="*60)
#         print(res.summary)

#         return res

#     except Exception as e:
#         print(f"ERROR in TWFE estimation: {e}")
#         print("\nDiagnostics:")
#         print(f"  Outcome var stats: {panel_df[outcome_var].describe()}")
#         print(f"  Treatment var stats: {panel_df[treatment_var].describe()}")
#         return None

# def estimate_event_study(panel_df, outcome_var, entity_var='grid_id', time_var='year',
#                         treatment_time_var='first_treatment_year', window=(-3, 5)):
#     """
#     Event study with leads and lags.
#     """
#     panel = panel_df.copy()

#     # Calculate event time (years relative to treatment)
#     panel['event_time'] = panel[time_var] - panel[treatment_time_var]

#     # Create event time dummies (omit -1 as reference)
#     for t in range(window[0], window[1] + 1):
#         if t == -1:
#             continue  # Reference period
#         panel[f'lead_lag_{t}'] = (panel['event_time'] == t).astype(int)

#     # Only include treated units and within window
#     panel_es = panel[panel[treatment_time_var].notna()].copy()
#     panel_es = panel_es[panel_es['event_time'].between(window[0], window[1])]

#     # Set index
#     panel_es = panel_es.set_index([entity_var, time_var])

#     # Build formula
#     lead_lag_vars = [f'lead_lag_{t}' for t in range(window[0], window[1] + 1) if t != -1]
#     formula = f"{outcome_var} ~ {' + '.join(lead_lag_vars)} + EntityEffects + TimeEffects"

#     try:
#         mod = PanelOLS.from_formula(formula, data=panel_es)
#         res = mod.fit(cov_type='clustered', cluster_entity=True)

#         print("Event Study Results")
#         print("="*60)
#         print(res.summary)

#         # Extract coefficients for plotting
#         coefs = []
#         for t in range(window[0], window[1] + 1):
#             if t == -1:
#                 coefs.append({'event_time': t, 'coef': 0, 'se': 0, 'ci_lower': 0, 'ci_upper': 0})
#             else:
#                 var_name = f'lead_lag_{t}'
#                 if var_name in res.params.index:
#                     coef = res.params[var_name]
#                     se = res.std_errors[var_name]
#                     ci_lower = coef - 1.96 * se
#                     ci_upper = coef + 1.96 * se
#                     coefs.append({'event_time': t, 'coef': coef, 'se': se, 'ci_lower': ci_lower, 'ci_upper': ci_upper})

#         coefs_df = pd.DataFrame(coefs)

#         return res, coefs_df
#     except Exception as e:
#         print(f"ERROR in event study: {e}")
#         return None, None

# print("DiD estimation functions defined (FIXED VERSION)")

# # 9. Difference-in-Differences estimation (FIXED VERSION)
# import pandas as pd
# import numpy as np
# from linearmodels.panel import PanelOLS
# from linearmodels import IV2SLS
# import statsmodels.api as sm
# from statsmodels.regression.linear_model import OLS

# def prepare_panel_data(outcome_df, treatment_df, grid_gdf):
#     """
#     Merge outcome and treatment into analysis panel.
#     """
#     # Merge treatment with outcomes
#     panel = outcome_df.merge(
#         treatment_df[['grid_id', 'year', 'treated', 'treatment_intensity', 'first_treatment_year']],
#         on=['grid_id', 'year'],
#         how='left'
#     )

#     # Fill missing treatment as control
#     panel['treated'] = panel['treated'].fillna(0)
#     panel['treatment_intensity'] = panel['treatment_intensity'].fillna(0)

#     # Add grid characteristics
#     panel = panel.merge(
#         grid_gdf[['grid_id', 'lon_center', 'lat_center', 'area_km2']],
#         on='grid_id',
#         how='left'
#     )

#     return panel

# def estimate_twfe_did_FIXED(panel_df, outcome_var, treatment_var='treated',
#                             entity_var='grid_id', time_var='year'):
#     """
#     FIXED: Two-way fixed effects DiD with proper error handling.
#     """
#     # Validate inputs
#     if panel_df[outcome_var].std() == 0:
#         print("ERROR: Outcome variable has zero variance!")
#         print(f"Unique values: {panel_df[outcome_var].unique()}")
#         return None

#     if panel_df[treatment_var].std() == 0:
#         print("WARNING: Treatment variable has zero variance!")
#         print(f"Unique values: {panel_df[treatment_var].unique()}")
#         return None

#     # Check for sufficient variation
#     n_treated = panel_df[treatment_var].sum()
#     n_total = len(panel_df)
#     print(f"Treatment prevalence: {n_treated}/{n_total} ({100*n_treated/n_total:.2f}%)")

#     # Set multi-index for panel
#     panel = panel_df.copy()
#     panel = panel.set_index([entity_var, time_var])

#     # Build formula
#     formula = f'{outcome_var} ~ {treatment_var} + EntityEffects + TimeEffects'

#     try:
#         mod = PanelOLS.from_formula(formula, data=panel)
#         res = mod.fit(cov_type='clustered', cluster_entity=True)

#         print("\nTwo-Way Fixed Effects DiD Results")
#         print("="*60)
#         print(res.summary)

#         return res

#     except Exception as e:
#         print(f"ERROR in TWFE estimation: {e}")
#         print("\nDiagnostics:")
#         print(f"  Outcome var stats: {panel_df[outcome_var].describe()}")
#         print(f"  Treatment var stats: {panel_df[treatment_var].describe()}")
#         return None
# def estimate_event_study(panel_df, outcome_var, entity_var='grid_id', time_var='year',
#                         treatment_time_var='first_treatment_year', window=(-3, 5)):
#     """
#     Event study with leads and lags. (FIXED VERSION 2 - Handles rank deficiency)
#     """
#     panel = panel_df.copy()

#     # Helper to create formula-safe names
#     def get_event_time_var(t):
#         if t < 0:
#             return f'lead_lag_m{abs(t)}'  # e.g., lead_lag_m3
#         else:
#             return f'lead_lag_{t}'       # e.g., lead_lag_0

#     # Calculate event time (years relative to treatment)
#     panel['event_time'] = panel[time_var] - panel[treatment_time_var]

#     # Create event time dummies (omit -1 as reference)
#     for t in range(window[0], window[1] + 1):
#         if t == -1:
#             continue  # Reference period
#         var_name = get_event_time_var(t)
#         panel[var_name] = (panel['event_time'] == t).astype(int)

#     # Only include treated units and within window
#     panel_es = panel[panel[treatment_time_var].notna()].copy()
#     panel_es = panel_es[panel_es['event_time'].between(window[0], window[1])]

#     # Set index
#     panel_es = panel_es.set_index([entity_var, time_var])

#     # --- START OF FIX ---
#     # Build formula *only with non-empty dummies*

#     # 1. Get all possible variable names from the window (except reference)
#     all_possible_vars = [get_event_time_var(t) for t in range(window[0], window[1] + 1) if t != -1]

#     # 2. Filter this list to include only variables that
#     #    (a) exist in our panel_es DataFrame AND
#     #    (b) have at least one observation (sum > 0)
#     lead_lag_vars = []
#     for var in all_possible_vars:
#         if var in panel_es.columns and panel_es[var].sum() > 0:
#             lead_lag_vars.append(var)

#     # 3. Check if we have any variables left to test
#     if not lead_lag_vars:
#         print("ERROR: No valid (non-empty) lead/lag variables found in the data window.")
#         print("This might be due to insufficient data for the treated group.")
#         return None, None

#     print(f"Running event study with {len(lead_lag_vars)} lead/lag variables.")

#     # 4. Build the formula string with the filtered list
#     formula = f"{outcome_var} ~ {' + '.join(lead_lag_vars)} + EntityEffects + TimeEffects"
#     # --- END OF FIX ---

#     try:
#         mod = PanelOLS.from_formula(formula, data=panel_es)
#         res = mod.fit(cov_type='clustered', cluster_entity=True)

#         print("Event Study Results")
#         print("=" * 60)
#         print(res.summary)

#         # Extract coefficients for plotting
#         coefs = []
#         for t in range(window[0], window[1] + 1):
#             if t == -1:
#                 # Add the reference period
#                 coefs.append({'event_time': t, 'coef': 0, 'se': 0, 'ci_lower': 0, 'ci_upper': 0})
#             else:
#                 var_name = get_event_time_var(t)

#                 # Check if this variable was in our final model
#                 if var_name in res.params.index:
#                     coef = res.params[var_name]
#                     se = res.std_errors[var_name]
#                     ci_lower = coef - 1.96 * se
#                     ci_upper = coef + 1.96 * se
#                     coefs.append({'event_time': t, 'coef': coef, 'se': se, 'ci_lower': ci_lower, 'ci_upper': ci_upper})
#                 # If it wasn't in the model (e.g., all zeros), don't add it
#                 # We can add it as 'None' or '0' if we want to see gaps in the plot
#                 elif var_name not in lead_lag_vars:
#                      # This var was all zeros, so we'll add it with NAs for plotting
#                      coefs.append({'event_time': t, 'coef': None, 'se': None, 'ci_lower': None, 'ci_upper': None})


#         coefs_df = pd.DataFrame(coefs).dropna(subset=['coef']) # Drop rows where coef is None

#         return res, coefs_df
#     except Exception as e:
#         print(f"ERROR in event study: {e}")
#         return None, None

# print("DiD estimation functions defined (FIXED VERSION)")

# 9. Difference-in-Differences estimation (FIXED VERSION 3)
import pandas as pd
import numpy as np
from linearmodels.panel import PanelOLS
from linearmodels import IV2SLS
import statsmodels.api as sm
from statsmodels.regression.linear_model import OLS

def prepare_panel_data(outcome_df, treatment_df, grid_gdf):
    """
    Merge outcome and treatment into analysis panel.
    """
    # Merge treatment with outcomes
    panel = outcome_df.merge(
        treatment_df[['grid_id', 'year', 'treated', 'treatment_intensity', 'first_treatment_year']],
        on=['grid_id', 'year'],
        how='left'
    )

    # Fill missing treatment as control
    panel['treated'] = panel['treated'].fillna(0)
    panel['treatment_intensity'] = panel['treatment_intensity'].fillna(0)

    # Add grid characteristics
    panel = panel.merge(
        grid_gdf[['grid_id', 'lon_center', 'lat_center', 'area_km2']],
        on='grid_id',
        how='left'
    )

    return panel

def estimate_twfe_did_FIXED(panel_df, outcome_var, treatment_var='treated',
                            entity_var='grid_id', time_var='year'):
    """
    FIXED: Two-way fixed effects DiD with proper error handling.
    """
    # Validate inputs
    if panel_df[outcome_var].std() == 0:
        print("ERROR: Outcome variable has zero variance!")
        print(f"Unique values: {panel_df[outcome_var].unique()}")
        return None

    if panel_df[treatment_var].std() == 0:
        print("WARNING: Treatment variable has zero variance!")
        print(f"Unique values: {panel_df[treatment_var].unique()}")
        return None

    # Check for sufficient variation
    n_treated = panel_df[treatment_var].sum()
    n_total = len(panel_df)
    print(f"Treatment prevalence: {n_treated}/{n_total} ({100*n_treated/n_total:.2f}%)")

    # Set multi-index for panel
    panel = panel_df.copy()
    panel = panel.set_index([entity_var, time_var])

    # Build formula
    formula = f'{outcome_var} ~ {treatment_var} + EntityEffects + TimeEffects'

    try:
        mod = PanelOLS.from_formula(formula, data=panel)
        res = mod.fit(cov_type='clustered', cluster_entity=True)

        print("\nTwo-Way Fixed Effects DiD Results")
        print("="*60)
        print(res.summary)

        return res

    except Exception as e:
        print(f"ERROR in TWFE estimation: {e}")
        print("\nDiagnostics:")
        print(f"  Outcome var stats: {panel_df[outcome_var].describe()}")
        print(f"  Treatment var stats: {panel_df[treatment_var].describe()}")
        return None

def estimate_event_study(panel_df, outcome_var, entity_var='grid_id', time_var='year',
                        treatment_time_var='first_treatment_year', window=(-3, 5)):
    """
    Event study with leads and lags. (FIXED VERSION 3 - Runs on FULL panel)
    """
    panel = panel_df.copy()

    # Helper to create formula-safe names
    def get_event_time_var(t):
        if t < 0:
            return f'lead_lag_m{abs(t)}'  # e.g., lead_lag_m3
        else:
            return f'lead_lag_{t}'       # e.g., lead_lag_0

    # Calculate event time (years relative to treatment)
    # This will be NaN for control units, which is correct.
    panel['event_time'] = panel[time_var] - panel[treatment_time_var]

    # Create event time dummies (omit -1 as reference)
    all_possible_vars = []
    for t in range(window[0], window[1] + 1):
        if t == -1:
            continue  # Reference period
        var_name = get_event_time_var(t)
        all_possible_vars.append(var_name)
        # For control units (event_time=NaN), (NaN == t) is False, so dummy is 0. This is correct.
        panel[var_name] = (panel['event_time'] == t).astype(int)

    # --- START OF FIX ---
    # We NO LONGER filter the panel. We run on the full panel.
    # panel_es = panel[panel[treatment_time_var].notna()].copy()  <-- REMOVED (INCORRECT)
    # panel_es = panel_es[panel_es['event_time'].between(window[0], window[1])] <-- REMOVED (INCORRECT)

    # We just need to set the index on the full panel
    panel = panel.set_index([entity_var, time_var])

    # Check which dummies are not all-zeros (to avoid rank deficiency)
    lead_lag_vars = []
    for var in all_possible_vars:
        if var in panel.columns and panel[var].sum() > 0:
            lead_lag_vars.append(var)

    if not lead_lag_vars:
        print("ERROR: No valid (non-empty) lead/lag variables found in the data window.")
        return None, None

    print(f"Running event study with {len(lead_lag_vars)} lead/lag variables on FULL panel.")

    # Build the formula string with the filtered list
    formula = f"{outcome_var} ~ {' + '.join(lead_lag_vars)} + EntityEffects + TimeEffects"
    # --- END OF FIX ---

    try:
        mod = PanelOLS.from_formula(formula, data=panel)
        res = mod.fit(cov_type='clustered', cluster_entity=True)

        print("Event Study Results")
        print("=" * 60)
        print(res.summary)

        # Extract coefficients for plotting
        coefs = []
        for t in range(window[0], window[1] + 1):
            if t == -1:
                # Add the reference period
                coefs.append({'event_time': t, 'coef': 0, 'se': 0, 'ci_lower': 0, 'ci_upper': 0})
            else:
                var_name = get_event_time_var(t)

                if var_name in res.params.index:
                    coef = res.params[var_name]
                    se = res.std_errors[var_name]
                    ci_lower = coef - 1.96 * se
                    ci_upper = coef + 1.96 * se
                    coefs.append({'event_time': t, 'coef': coef, 'se': se, 'ci_lower': ci_lower, 'ci_upper': ci_upper})
                elif var_name in all_possible_vars:
                     # This var was all zeros and dropped, add Nones
                     coefs.append({'event_time': t, 'coef': None, 'se': None, 'ci_lower': None, 'ci_upper': None})

        # Drop rows where coef is None (from all-zero dummies)
        coefs_df = pd.DataFrame(coefs).dropna(subset=['coef'])

        return res, coefs_df
    except Exception as e:
        print(f"ERROR in event study: {e}")
        return None, None

print("DiD estimation functions defined (FIXED VERSION 3)")
```

```python
# 13. Robustness check implementations
import numpy as np
import pandas as pd
from linearmodels.panel import PanelOLS

def pretrend_test(event_study_coefs, window=(-3, -1)):
    """
    Test for parallel trends: joint F-test on pre-treatment coefficients.
    """
    pre_coefs = event_study_coefs[
        event_study_coefs['event_time'].between(window[0], window[1])
    ]

    # Joint test: are all pre-treatment coefs jointly zero?
    # Use F-statistic
    mean_coef = pre_coefs['coef'].mean()
    se = pre_coefs['se'].mean()  # Simplified

    f_stat = (mean_coef / se) ** 2

    print("Pre-Trend Test")
    print("=" * 40)
    print(f"Mean pre-treatment coefficient: {mean_coef:.4f}")
    print(f"F-statistic (simplified): {f_stat:.4f}")
    print("\nInterpretation: Small F-stat and coefs close to 0 support parallel trends")

    return f_stat

def placebo_test(panel_df, outcome_var, n_iterations=100, seed=42):
    """
    Placebo test: randomly assign treatment and estimate effects.
    """
    np.random.seed(seed)

    placebo_coefs = []

    for i in range(n_iterations):
        # Shuffle treatment assignment
        panel_placebo = panel_df.copy()
        panel_placebo['treated'] = np.random.permutation(panel_placebo['treated'].values)

        # Estimate
        try:
            panel_placebo_indexed = panel_placebo.set_index(['grid_id', 'year'])
            mod = PanelOLS.from_formula(
                f'{outcome_var} ~ treated + EntityEffects + TimeEffects',
                data=panel_placebo_indexed
            )
            res = mod.fit(cov_type='clustered', cluster_entity=True)
            placebo_coefs.append(res.params['treated'])
        except:
            continue

    placebo_coefs = np.array(placebo_coefs)

    print("Placebo Test Results")
    print("=" * 40)
    print(f"Mean placebo coefficient: {placebo_coefs.mean():.4f}")
    print(f"Std dev: {placebo_coefs.std():.4f}")
    print(f"95% CI: [{np.percentile(placebo_coefs, 2.5):.4f}, {np.percentile(placebo_coefs, 97.5):.4f}]")

    return placebo_coefs

def sensitivity_alternative_resolution(occurrence_df, treatment_df, resolutions=[0.05, 0.1, 0.2]):
    """
    Re-run analysis at different spatial resolutions.
    """
    results = {}

    for res in resolutions:
        print(f"\n{'='*60}")
        print(f"Analyzing at {res}° resolution...")

        # Create grid at this resolution
        grid = create_analysis_grid(CONFIG['region_bbox'], res)

        # Re-aggregate data
        # ... (implementation similar to main analysis)

        results[res] = {'grid': grid}  # Store results

    return results

print("Robustness check functions defined")
```

```python
# Step 6: Species-Specific Analysis
print("\n" + "="*80)
print("STEP 6: Species-Specific Treatment Effects")
print("="*80)

if 'panel_df' in locals():
    species_results = {}

    # Get list of species with sufficient data
    species_counts = panel_df.groupby('species').size()
    species_to_analyze = species_counts[species_counts >= CONFIG['min_occurrences']].index.tolist()

    print(f"Analyzing {len(species_to_analyze)} species with ≥{CONFIG['min_occurrences']} observations")

    for i, species in enumerate(species_to_analyze, 1):
        print(f"\n[{i}/{len(species_to_analyze)}] Analyzing {species}...")

        # Filter panel to this species
        panel_sp = panel_df[panel_df['species'] == species].copy()

        # Check if there's variation in treatment
        if panel_sp['treated'].sum() == 0:
            print(f"  Skipping - no treated observations")
            continue

        try:
            res = estimate_twfe_did_FIXED(  # Changed from estimate_twfe_did
                panel_sp,
                outcome_var='occupancy',
                treatment_var='treated'
            )

            if res is not None:
                species_results[species] = {
                    'coef': res.params['treated'],
                    'se': res.std_errors['treated'],
                    'ci_lower': res.conf_int().loc['treated', 'lower'],
                    'ci_upper': res.conf_int().loc['treated', 'upper'],
                    'n_obs': len(panel_sp),
                    'n_treated': panel_sp['treated'].sum()
                }
                print(f"  Effect: {species_results[species]['coef']:.4f} (SE: {species_results[species]['se']:.4f})")
        except Exception as e:
            print(f"  ERROR: {str(e)[:100]}")
            continue

    print(f"\n{'='*60}")
    print(f"Successfully analyzed {len(species_results)} species")

    # Save results
    species_results_df = pd.DataFrame(species_results).T
    species_results_df.index.name = 'species'
    species_results_path = CONFIG['results_dir'] / 'species_specific_results.csv'
    species_results_df.to_csv(species_results_path)
    print(f"Saved to {species_results_path}")
else:
    print("ERROR: Analysis panel not found")
```

```python
# # Advanced methods cell needs complete rewrite
# import numpy as np
# import pandas as pd
# from linearmodels.panel import PanelOLS
# from scipy.optimize import minimize

# def estimate_sun_abraham(panel_df, outcome_var, entity_var='grid_id',
#                         time_var='year', treatment_time_var='first_treatment_year'):
#     """
#     Sun-Abraham (2021) interaction-weighted estimator for staggered DiD.

#     Handles heterogeneous treatment effects and staggered adoption.
#     More robust than TWFE when treatment effects vary over time.
#     """
#     print("\n" + "="*60)
#     print("Sun-Abraham (2021) Estimator")
#     print("="*60)

#     panel = panel_df.copy()

#     # Create cohort indicators (year first treated)
#     panel['cohort'] = panel[treatment_time_var]
#     panel['cohort'] = panel['cohort'].fillna(9999)  # Never-treated = 9999

#     # Create relative time indicators
#     panel['rel_time'] = panel[time_var] - panel[treatment_time_var]
#     panel.loc[panel['cohort'] == 9999, 'rel_time'] = -999  # Never-treated

#     # Interaction-weighted estimation
#     results = {}
#     cohorts = panel[panel['cohort'] != 9999]['cohort'].unique()

#     for cohort in sorted(cohorts):
#         print(f"\nEstimating for cohort {int(cohort)}...")

#         # For each cohort, compare to not-yet-treated units
#         cohort_data = panel[
#             (panel['cohort'] == cohort) |
#             (panel['cohort'] > cohort) |
#             (panel['cohort'] == 9999)
#         ].copy()

#         # Create treatment indicator
#         cohort_data['treated_cohort'] = (
#             (cohort_data['cohort'] == cohort) &
#             (cohort_data[time_var] >= cohort)
#         ).astype(int)

#         # Estimate
#         cohort_data_indexed = cohort_data.set_index([entity_var, time_var])

#         try:
#             mod = PanelOLS.from_formula(
#                 f'{outcome_var} ~ treated_cohort + EntityEffects + TimeEffects',
#                 data=cohort_data_indexed
#             )
#             res = mod.fit(cov_type='clustered', cluster_entity=True)

#             results[int(cohort)] = {
#                 'coef': res.params['treated_cohort'],
#                 'se': res.std_errors['treated_cohort'],
#                 'n_treated': cohort_data['treated_cohort'].sum(),
#                 'n_control': len(cohort_data) - cohort_data['treated_cohort'].sum()
#             }

#             print(f"  ATT = {results[int(cohort)]['coef']:.4f} (SE: {results[int(cohort)]['se']:.4f})")

#         except Exception as e:
#             print(f"  ERROR: {e}")
#             continue

#     # Compute overall ATT (weighted average)
#     if results:
#         weights = np.array([r['n_treated'] for r in results.values()])
#         weights = weights / weights.sum()

#         overall_att = np.sum([r['coef'] * w for r, w in zip(results.values(), weights)])
#         overall_se = np.sqrt(np.sum([r['se']**2 * w**2 for r, w in zip(results.values(), weights)]))

#         print("\n" + "="*60)
#         print("OVERALL ATT (Sun-Abraham)")
#         print("="*60)
#         print(f"Coefficient: {overall_att:.4f}")
#         print(f"Std Error: {overall_se:.4f}")
#         print(f"95% CI: [{overall_att - 1.96*overall_se:.4f}, {overall_att + 1.96*overall_se:.4f}]")

#         return {
#             'overall_att': overall_att,
#             'overall_se': overall_se,
#             'cohort_results': results
#         }

#     return None

# def estimate_synthetic_control(panel_df, treated_unit, outcome_var,
#                                treatment_time, entity_var='grid_id', time_var='year'):
#     """
#     Synthetic Control Method (Abadie et al. 2010).

#     Creates a weighted combination of control units that best matches
#     the treated unit in pre-treatment period.

#     Args:
#         treated_unit: ID of the treated unit
#         treatment_time: Year of treatment
#     """
#     print("\n" + "="*60)
#     print("Synthetic Control Method")
#     print("="*60)

#     # Separate treated and control units
#     treated = panel_df[panel_df[entity_var] == treated_unit].copy()
#     controls = panel_df[panel_df[entity_var] != treated_unit].copy()

#     # Pre-treatment period
#     pre_treatment_years = treated[treated[time_var] < treatment_time][time_var].unique()

#     print(f"Treated unit: {treated_unit}")
#     print(f"Treatment time: {treatment_time}")
#     print(f"Pre-treatment periods: {len(pre_treatment_years)}")
#     print(f"Available control units: {controls[entity_var].nunique()}")

#     # Get pre-treatment outcomes for treated unit
#     Y1_pre = treated[treated[time_var] < treatment_time][outcome_var].values

#     # Get pre-treatment outcomes for all control units
#     control_units = controls[entity_var].unique()
#     Y0_pre = np.zeros((len(pre_treatment_years), len(control_units)))

#     for i, unit in enumerate(control_units):
#         unit_data = controls[
#             (controls[entity_var] == unit) &
#             (controls[time_var] < treatment_time)
#         ].sort_values(time_var)
#         Y0_pre[:, i] = unit_data[outcome_var].values

#     # Optimization: find weights that minimize pre-treatment prediction error
#     def objective(w):
#         synthetic = Y0_pre @ w
#         return np.sum((Y1_pre - synthetic) ** 2)

#     def constraint_sum(w):
#         return np.sum(w) - 1

#     # Constraints: weights sum to 1 and are non-negative
#     constraints = {'type': 'eq', 'fun': constraint_sum}
#     bounds = [(0, 1) for _ in range(len(control_units))]

#     # Initial guess: equal weights
#     w0 = np.ones(len(control_units)) / len(control_units)

#     # Optimize
#     result = minimize(objective, w0, method='SLSQP', bounds=bounds, constraints=constraints)

#     if not result.success:
#         print("WARNING: Optimization did not converge")
#         return None

#     weights = result.x

#     # Units with non-trivial weights
#     significant_weights = weights[weights > 0.01]
#     print(f"\nUnits in synthetic control: {len(significant_weights)}")
#     print(f"Top 5 weights: {np.sort(weights)[-5:][::-1]}")

#     # Construct synthetic control for all periods
#     synthetic_outcomes = []
#     actual_outcomes = []
#     years = sorted(treated[time_var].unique())

#     for year in years:
#         # Actual treated outcome
#         actual = treated[treated[time_var] == year][outcome_var].values[0]
#         actual_outcomes.append(actual)

#         # Synthetic control outcome
#         control_outcomes = []
#         for unit in control_units:
#             unit_outcome = controls[
#                 (controls[entity_var] == unit) &
#                 (controls[time_var] == year)
#             ][outcome_var].values
#             if len(unit_outcome) > 0:
#                 control_outcomes.append(unit_outcome[0])
#             else:
#                 control_outcomes.append(0)

#         synthetic = np.dot(weights, control_outcomes)
#         synthetic_outcomes.append(synthetic)

#     # Compute treatment effects
#     effects = np.array(actual_outcomes) - np.array(synthetic_outcomes)

#     # Post-treatment average effect
#     post_periods = [i for i, year in enumerate(years) if year >= treatment_time]
#     att = np.mean(effects[post_periods])

#     print(f"\nAverage Treatment Effect on Treated (ATT): {att:.4f}")
#     print(f"Pre-treatment RMSPE: {np.sqrt(np.mean(effects[:len(pre_treatment_years)]**2)):.4f}")
#     print(f"Post-treatment RMSPE: {np.sqrt(np.mean(effects[post_periods]**2)):.4f}")

#     return {
#         'weights': weights,
#         'control_units': control_units,
#         'years': years,
#         'actual': actual_outcomes,
#         'synthetic': synthetic_outcomes,
#         'effects': effects,
#         'att': att,
#         'treatment_time': treatment_time
#     }

# def plot_synthetic_control(sc_results, save_path=None):
#     """Plot synthetic control results."""
#     import matplotlib.pyplot as plt

#     fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

#     years = sc_results['years']
#     treatment_time = sc_results['treatment_time']

#     # Panel A: Trends
#     ax1.plot(years, sc_results['actual'], 'o-', label='Treated', linewidth=2)
#     ax1.plot(years, sc_results['synthetic'], 's--', label='Synthetic Control', linewidth=2)
#     ax1.axvline(treatment_time, color='red', linestyle='--', alpha=0.5, label='Treatment')
#     ax1.set_xlabel('Year')
#     ax1.set_ylabel('Outcome')
#     ax1.set_title('Treated vs Synthetic Control')
#     ax1.legend()
#     ax1.grid(alpha=0.3)

#     # Panel B: Gap (treatment effect)
#     ax2.plot(years, sc_results['effects'], 'o-', color='darkgreen', linewidth=2)
#     ax2.axhline(0, color='black', linestyle='-', linewidth=1)
#     ax2.axvline(treatment_time, color='red', linestyle='--', alpha=0.5)
#     ax2.fill_between(years, 0, sc_results['effects'], alpha=0.3, color='green')
#     ax2.set_xlabel('Year')
#     ax2.set_ylabel('Treatment Effect')
#     ax2.set_title('Gap: Treated - Synthetic Control')
#     ax2.grid(alpha=0.3)

#     plt.tight_layout()

#     if save_path:
#         plt.savefig(save_path, dpi=300, bbox_inches='tight')
#     plt.show()

# def placebo_in_space(panel_df, outcome_var, treatment_df, n_iterations=100):
#     """
#     Spatial placebo test: randomly reassign which grid cells were treated.

#     If true effect, randomized treatment should have no effect.
#     """
#     print("\nRunning spatial placebo test...")

#     placebo_effects = []

#     for i in range(n_iterations):
#         # Shuffle treatment assignment across grid cells
#         panel_placebo = panel_df.copy()

#         # Get unique treated grid cells
#         treated_grids = treatment_df['grid_id'].unique()
#         all_grids = panel_df['grid_id'].unique()

#         # Randomly select same number of grid cells
#         np.random.seed(i)
#         placebo_treated = np.random.choice(all_grids, size=len(treated_grids), replace=False)

#         # Reassign treatment
#         panel_placebo['treated'] = panel_placebo['grid_id'].isin(placebo_treated).astype(int)

#         # Estimate
#         try:
#             res = estimate_twfe_did_FIXED(panel_placebo, outcome_var, treatment_var='treated')
#             if res is not None:
#                 placebo_effects.append(res.params['treated'])
#         except:
#             continue

#     placebo_effects = np.array(placebo_effects)

#     print(f"Placebo effects (n={len(placebo_effects)})")
#     print(f"  Mean: {placebo_effects.mean():.4f}")
#     print(f"  Std: {placebo_effects.std():.4f}")
#     print(f"  95% CI: [{np.percentile(placebo_effects, 2.5):.4f}, {np.percentile(placebo_effects, 97.5):.4f}]")

#     return placebo_effects

# print("Advanced estimation methods defined")
# Advanced estimation methods defined (FIXED SCM PERFORMANCE v2)
import numpy as np
import pandas as pd
from linearmodels.panel import PanelOLS
from scipy.optimize import minimize

def estimate_sun_abraham(panel_df, outcome_var, entity_var='grid_id',
                        time_var='year', treatment_time_var='first_treatment_year'):
    """
    Sun-Abraham (2021) estimator.
    """
    print("\n" + "="*60)
    print("Sun-Abraham (2021) Estimator")
    print("="*60)
    panel = panel_df.copy()
    panel['cohort'] = panel[treatment_time_var]
    panel['cohort'] = panel['cohort'].fillna(9999)  # Never-treated = 9999
    panel['rel_time'] = panel[time_var] - panel[treatment_time_var]
    panel.loc[panel['cohort'] == 9999, 'rel_time'] = -999  # Never-treated
    results = {}
    cohorts = panel[panel['cohort'] != 9999]['cohort'].unique()
    for cohort in sorted(cohorts):
        if pd.isna(cohort):
            continue
        print(f"\nEstimating for cohort {int(cohort)}...")
        cohort_data = panel[
            (panel['cohort'] == cohort) |
            (panel['cohort'] > cohort) |
            (panel['cohort'] == 9999)
        ].copy()
        cohort_data['treated_cohort'] = (
            (cohort_data['cohort'] == cohort) &
            (cohort_data[time_var] >= cohort)
        ).astype(int)
        cohort_data_indexed = cohort_data.set_index([entity_var, time_var])
        try:
            mod = PanelOLS.from_formula(
                f'{outcome_var} ~ treated_cohort + EntityEffects + TimeEffects',
                data=cohort_data_indexed
            )
            res = mod.fit(cov_type='clustered', cluster_entity=True)
            results[int(cohort)] = {
                'coef': res.params['treated_cohort'],
                'se': res.std_errors['treated_cohort'],
                'n_treated': cohort_data['treated_cohort'].sum(),
                'n_control': len(cohort_data) - cohort_data['treated_cohort'].sum()
            }
            print(f"  ATT = {results[int(cohort)]['coef']:.4f} (SE: {results[int(cohort)]['se']:.4f})")
        except Exception as e:
            print(f"  ERROR: {e}")
            continue
    if results:
        valid_results = [r for r in results.values() if r['n_treated'] > 0]
        if not valid_results:
             print("ERROR: No treated units found in any cohort.")
             return None
        weights = np.array([r['n_treated'] for r in valid_results])
        weights = weights / weights.sum()
        coefs = np.array([r['coef'] for r in valid_results])
        ses = np.array([r['se']**2 for r in valid_results])
        overall_att = np.sum(coefs * weights)
        overall_se = np.sqrt(np.sum(ses * weights**2))
        print("\n" + "="*60)
        print("OVERALL ATT (Sun-Abraham)")
        print("="*60)
        print(f"Coefficient: {overall_att:.4f}")
        print(f"Std Error: {overall_se:.4f}")
        print(f"95% CI: [{overall_att - 1.96*overall_se:.4f}, {overall_att + 1.96*overall_se:.4f}]")
        return {
            'overall_att': overall_att,
            'overall_se': overall_se,
            'cohort_results': results
        }
    return None

def estimate_twfe_did_FIXED(panel_df, outcome_var, treatment_var='treated',
                            entity_var='grid_id', time_var='year',
                            _drop_absorbed=False):
    """
    FIXED: Two-way fixed effects DiD with proper error handling.
    Now accepts _drop_absorbed.
    """
    if panel_df[outcome_var].std() == 0:
        return None
    if panel_df[treatment_var].std() == 0:
        return None
    panel = panel_df.copy()
    panel = panel.set_index([entity_var, time_var])
    formula = f'{outcome_var} ~ {treatment_var} + EntityEffects + TimeEffects'
    try:
        mod = PanelOLS.from_formula(formula, data=panel)
        res = mod.fit(cov_type='clustered', cluster_entity=True,
                      drop_absorbed=_drop_absorbed)
        return res
    except Exception as e:
        raise e

# --- START OF SCM FIX (PERFORMANCE v2) ---
# This function is now FAST and uses vectorized pandas operations
# NEW: Added n_controls_sample parameter
def estimate_synthetic_control(panel_df, treated_unit, outcome_var,
                               treatment_time, entity_var='grid_id', time_var='year',
                               n_controls_sample=1000, random_seed=42): # <-- NEW PARAMETER
    """
    Synthetic Control Method (Abadie et al. 2010).
    (FIXED FOR PERFORMANCE - uses pivot_table, no loops)
    (NOW WITH SAMPLING for speed)
    """
    print("\n" + "="*60)
    print("Synthetic Control Method (High Performance)")
    print("="*60)

    # Separate treated and control units
    treated_df = panel_df[panel_df[entity_var] == treated_unit].copy()
    controls_df = panel_df[panel_df[entity_var] != treated_unit].copy()

    # Filter controls to only NEVER-TREATED units
    never_treated_ids = controls_df[controls_df['first_treatment_year'].isna()][entity_var].unique()

    # --- NEW: SAMPLING LOGIC ---
    if n_controls_sample is not None and len(never_treated_ids) > n_controls_sample:
        print(f"Sampling {n_controls_sample} random control units (from {len(never_treated_ids)}) for speed.")
        np.random.seed(random_seed)
        never_treated_ids = np.random.choice(never_treated_ids, n_controls_sample, replace=False)
    # --- END SAMPLING LOGIC ---

    controls_df = controls_df[controls_df[entity_var].isin(never_treated_ids)]

    if len(controls_df) == 0:
        print("ERROR: No never-treated control units found for SCM.")
        return None

    # Pre-treatment period
    pre_treatment_years = sorted(treated_df[treated_df[time_var] < treatment_time][time_var].unique())

    if len(pre_treatment_years) == 0:
        print(f"ERROR: Treated unit {treated_unit} has no pre-treatment periods before {treatment_time}.")
        return None

    print(f"Treated unit: {treated_unit}")
    print(f"Treatment time: {treatment_time}")
    print(f"Pre-treatment periods: {len(pre_treatment_years)} (from {min(pre_treatment_years)} to {max(pre_treatment_years)})")
    print(f"Available control units: {controls_df[entity_var].nunique()}")

    # --- Vectorized Data Prep ---
    print("Pivoting control data (this is fast)...")

    # 1. Get pre-treatment outcomes for treated unit
    Y1_pre_df = treated_df[treated_df[time_var].isin(pre_treatment_years)].set_index(time_var)[outcome_var].sort_index()
    Y1_pre = Y1_pre_df.values

    # 2. Pivot the control data: grid_id as index, year as columns
    controls_pre = controls_df[controls_df[time_var].isin(pre_treatment_years)]
    control_pivot = controls_pre.pivot_table(
        index=entity_var,
        columns=time_var,
        values=outcome_var
    )

    # 3. Drop any control units that have missing data in the pre-period
    control_pivot = control_pivot.dropna()

    # 4. Get the final numpy array (Y0_pre) and the list of valid units
    if control_pivot.empty:
        print("ERROR: No control units have full pre-treatment data.")
        return None

    # Align columns with Y1_pre_df's index (sorted years)
    control_pivot = control_pivot[Y1_pre_df.index]

    Y0_pre = control_pivot.values.T  # Transpose to get (Time x Units)
    valid_control_units = control_pivot.index.tolist()

    print(f"Using {len(valid_control_units)} control units with full data.")
    # --- End Vectorized Data Prep ---


    # Optimization: find weights that minimize pre-treatment prediction error
    def objective(w):
        synthetic = Y0_pre @ w
        return np.sum((Y1_pre - synthetic) ** 2)

    def constraint_sum(w):
        return np.sum(w) - 1

    # Constraints: weights sum to 1 and are non-negative
    constraints = {'type': 'eq', 'fun': constraint_sum}
    bounds = [(0, 1) for _ in range(len(valid_control_units))]

    # Initial guess: equal weights
    w0 = np.ones(len(valid_control_units)) / len(valid_control_units)

    # Optimize
    print("Optimizing weights...")
    result = minimize(objective, w0, method='SLSQP', bounds=bounds, constraints=constraints)

    if not result.success:
        print("WARNING: Optimization did not converge")
        return None

    weights = result.x

    # Units with non-trivial weights
    significant_weights = weights[weights > 0.01]
    print(f"\nUnits in synthetic control: {len(significant_weights)}")
    print(f"Top 5 weights: {np.sort(weights)[-5:][::-1]}")

    # --- Vectorized Outcome Construction ---
    print("Constructing synthetic outcomes...")

    # 1. Get *all* data for the valid control units
    controls_full = controls_df[controls_df[entity_var].isin(valid_control_units)]

    # 2. Pivot the *full* timeline
    controls_full_pivot = controls_full.pivot_table(
        index=entity_var,
        columns=time_var,
        values=outcome_var
    )

    # 3. Re-order rows to match the order of 'weights' and fill missing (post-treatment) with 0
    controls_full_ordered = controls_full_pivot.reindex(valid_control_units).fillna(0)

    # 4. Get the (Time x Units) data
    Y0_full_df = controls_full_ordered.T

    # 5. Get the actual (Y1) data
    Y1_full_df = treated_df.set_index(time_var)[outcome_var]

    # 6. Align them (in case of missing years in one or the other)
    combined_df = Y1_full_df.to_frame(name='actual').join(Y0_full_df, how='inner')

    years = combined_df.index.values
    actual_outcomes = combined_df['actual'].values
    Y0_aligned_values = combined_df[valid_control_units].values # (Years x Units)

    # 7. Calculate synthetic control with one matrix multiplication
    synthetic_outcomes = Y0_aligned_values @ weights
    # --- End Vectorized Outcome Construction ---

    # Compute treatment effects
    effects = actual_outcomes - synthetic_outcomes

    # Post-treatment average effect
    post_periods = (years >= treatment_time)
    if not post_periods.any():
        print("Warning: No post-treatment periods found.")
        att = 0
    else:
        att = np.mean(effects[post_periods])

    print(f"\nAverage Treatment Effect on Treated (ATT): {att:.4f}")

    pre_periods = (years < treatment_time)
    if pre_periods.any():
        print(f"Pre-treatment RMSPE: {np.sqrt(np.mean(effects[pre_periods]**2)):.4f}")
    if post_periods.any():
        print(f"Post-treatment RMSPE: {np.sqrt(np.mean(effects[post_periods]**2)):.4f}")

    return {
        'weights': weights,
        'control_units': valid_control_units,
        'years': years,
        'actual': actual_outcomes,
        'synthetic': synthetic_outcomes,
        'effects': effects,
        'att': att,
        'treatment_time': treatment_time
    }
# --- END OF SCM FIX ---


def plot_synthetic_control(sc_results, save_path=None):
    """Plot synthetic control results."""
    import matplotlib.pyplot as plt

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

    years = sc_results['years']
    treatment_time = sc_results['treatment_time']

    # Panel A: Trends
    ax1.plot(years, sc_results['actual'], 'o-', label='Treated', linewidth=2)
    ax1.plot(years, sc_results['synthetic'], 's--', label='Synthetic Control', linewidth=2)
    ax1.axvline(treatment_time, color='red', linestyle='--', alpha=0.5, label='Treatment')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Outcome')
    ax1.set_title('Treated vs Synthetic Control')
    ax1.legend()
    ax1.grid(alpha=0.3)

    # Panel B: Gap (treatment effect)
    ax2.plot(years, sc_results['effects'], 'o-', color='darkgreen', linewidth=2)
    ax2.axhline(0, color='black', linestyle='-', linewidth=1)
    ax2.axvline(treatment_time, color='red', linestyle='--', alpha=0.5)
    ax2.fill_between(years, 0, sc_results['effects'], alpha=0.3, color='green')
    ax2.set_xlabel('Year')
    ax2.set_ylabel('Treatment Effect')
    ax2.set_title('Gap: Treated - Synthetic Control')
    ax2.grid(alpha=0.3)

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()

def placebo_in_space(panel_df, outcome_var, entity_var='grid_id', time_var='year', n_iterations=100, seed=42):
    """
    Placebo test (FIXED): Randomly shuffle treatment assignment (the 'first_treatment_year')
    across all grid cells and re-run the DiD.
    Now with fallback logic for absorbed variables.
    """
    print("\nRunning spatial placebo test (shuffling treatment cohorts)...")
    np.random.seed(seed)

    placebo_coefs = []

    # Get the universe of grid IDs
    all_grid_ids = panel_df[entity_var].unique()

    # Get the actual distribution of treatment cohorts (first_treatment_year)
    # This includes NaT for never-treated units
    cohort_distribution = panel_df.drop_duplicates(subset=[entity_var])[['first_treatment_year']]

    # We must have the estimate_twfe_did_FIXED function available in the global scope
    if 'estimate_twfe_did_FIXED' not in globals():
        print("ERROR: estimate_twfe_did_FIXED function not found.")
        return np.array([])

    for i in range(n_iterations):
        if (i+1) % 10 == 0:
            print(f"  ... running iteration {i+1}/{n_iterations}")

        panel_placebo = panel_df.copy()

        # --- Start of Shuffling Logic ---
        # Shuffle the cohort distribution
        shuffled_cohorts_values = cohort_distribution.sample(frac=1).values.flatten()

        # Create a mapping from grid_id to its new placebo cohort
        placebo_cohort_map = pd.DataFrame({
            entity_var: all_grid_ids,
            'placebo_first_year': shuffled_cohorts_values
        })

        # Merge this map back into the main panel
        if 'treated' in panel_placebo.columns:
            panel_placebo = panel_placebo.drop(columns=['treated']) # Drop old 'treated'
        panel_placebo = panel_placebo.merge(placebo_cohort_map, on=entity_var, how='left')

        # Re-create the 'treated' variable based on the *placebo* treatment year
        panel_placebo['treated'] = (panel_placebo[time_var] >= panel_placebo['placebo_first_year']).astype(int)
        # --- End of Shuffling Logic ---

        # Estimate DiD with the new placebo 'treated' variable
        try:
            # First, try the strict model
            res = estimate_twfe_did_FIXED(
                panel_placebo,
                outcome_var,
                treatment_var='treated',
                entity_var=entity_var,
                time_var=time_var,
                _drop_absorbed=False # <-- Try strict first
            )
            if res is not None:
                placebo_coefs.append(res.params['treated'])
        except Exception as e:
            # If it fails (likely absorption), try again with drop_absorbed=True
            try:
                res = estimate_twfe_did_FIXED(
                    panel_placebo,
                    outcome_var,
                    treatment_var='treated',
                    entity_var=entity_var,
                    time_var=time_var,
                    _drop_absorbed=True # <-- Loosen the restriction
                )
                if res is not None:
                    # Check if 'treated' was dropped
                    if 'treated' in res.params:
                        placebo_coefs.append(res.params['treated'])
                    else:
                        pass # 'treated' was absorbed, but the model ran
            except Exception as e2:
                # This iteration failed completely
                continue

    placebo_coefs = np.array(placebo_coefs)

    if len(placebo_coefs) == 0:
        print("ERROR: All placebo iterations failed.")
        return np.array([])

    print(f"\nPlacebo effects (n={len(placebo_coefs)})")
    print(f"  Mean: {placebo_coefs.mean():.4f}")
    print(f"  Std: {placebo_coefs.std():.4f}")
    print(f"  95% CI: [{np.percentile(placebo_coefs, 2.5):.4f}, {np.percentile(placebo_coefs, 97.5):.4f}]")

    return placebo_coefs


print("Advanced estimation methods defined (FIXED SCM PERFORMANCE v2)")
```
